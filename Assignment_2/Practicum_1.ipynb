## 1. Smoothing

##### a. Average Filter: Apply an averaging filter of kernel size 10X10 on the original image.

Import the required libraries

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

Read the original image and convert them into gray scale and display them

```python
img_color = cv2.imread('image-1.jpg', 1)
img_color = img_color[:, :, ::-1]

img_gray = cv2.imread('image-1.jpg', 0)

plt.subplot(121), plt.imshow(img_color)
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(img_gray, cmap='gray')
plt.title('Grey scale'), plt.xticks([]), plt.yticks([])

plt.show()
```

Creating the 10x10 kernel of the average filter

```python
kernel = np.ones((10, 10), np.float32)/100
```

Putting a negative value for depth to indicate to use the same depth of the source image, the Filter2D operation convolves the image with the kernel.

```python
smooth_dst_color = cv2.filter2D(img_color, -1, kernel)
smooth_dst_gray = cv2.filter2D(img_gray, -1, kernel)
```

Display the original image, gray scale and their smoothened images in subplot

```python
plt.subplot(221), plt.imshow(img_color)
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(222), plt.imshow(img_gray, cmap='gray')
plt.title('Gray Scale'), plt.xticks([]), plt.yticks([])
plt.subplot(223), plt.imshow(smooth_dst_color)
plt.title('Smoothened Color'), plt.xticks([]), plt.yticks([])
plt.subplot(224), plt.imshow(smooth_dst_gray, cmap='gray')
plt.title('Smoothened Gray'), plt.xticks([]), plt.yticks([])

plt.show()
```

# Task B

## 1. Smoothing

##### a. Apply a smoothening filter on the original image

The idea of Task B 1 is come from websit: https://docs.opencv.org/master/de/dbc/tutorial_py_fourier_transform.html

Import the required libraries

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

Read the image and convert it into gray scale and convert the gray scale into float32

```python
img = cv2.imread('image-1.jpg',0)
img_float32 = np.float32(img)
```

Use function np.fft.fft2 compute the 2-dimensional discrete Fourier Transform. Use np.fft.fftshift to move the zero-frequency component to the center of the array. Calculate the magnitude spectrum

```python
dft = cv2.dft(img_float32, flags = cv2.DFT_COMPLEX_OUTPUT)
dft_shift = np.fft.fftshift(dft)
magnitude_spectrum = 20*np.log(cv2.magnitude(dft_shift[:,:,0],dft_shift[:,:,1]))
```

Get the columns and rows of the gray scale

```python
rows, cols = img.shape
crow, ccol = rows//2 , cols//2 
```

Create a mask which the center squancre is 1, and remaining all zeros

```python
mask = np.zeros((rows, cols, 2), np.uint8)
mask[crow-55:crow+55, ccol-55:ccol+55] = 1
```

Imply the mask

```python
fshift = dft_shift*mask
f_ishift = np.fft.ifftshift(fshift)
```

Perform the inverse DFT

```python
img_back = cv2.idft(f_ishift)
img_back = cv2.magnitude(img_back[:,:,0],img_back[:,:,1])
```

Display the gray scale, magnitude spectrum and the smoothened image in subplot

```python
plt.subplot(131),plt.imshow(img, cmap = 'gray')
plt.title('Gray Scale'), plt.xticks([]), plt.yticks([])
plt.subplot(132),plt.imshow(magnitude_spectrum, cmap = 'gray')
plt.title('Magnitude Spectrum'), plt.xticks([]), plt.yticks([])
plt.subplot(133),plt.imshow(img_back, cmap = 'gray')
plt.title('Smoothened Gray'), plt.xticks([]), plt.yticks([])

plt.show()
```

## 2. Edge Detection

##### a. Apply an Edge Detection algorithm on the original image

The idea of Task B 2 is come from website: https://www.programmersought.com/article/15934711254/

Import the required libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

Read the image and convert it into gray scale

```python
img = cv2.imread('image-1.jpg',0)
```

Use function np.fft.fft2 compute the 2-dimensional discrete Fourier Transform.
Use np.fft.fftshift to move the zero-frequency component to the center of the array.
Calculate the magnitude spectrum

```python
f = np.fft.fft2(img)
fshift = np.fft.fftshift(f)
magnitude_spectrum = 20*np.log(np.abs(fshift))
```

Get the columns and rows of the gray scale

```python
rows,cols=img.shape
crow,ccol=int(rows/2),int(cols/2)
```

Imply the Low frequency filter

```python
fshift[(crow-30):(crow+30),(ccol-30):(ccol+30)]=0
```

Perform the Inverse Transformation

```python
f_ishift=np.fft.ifftshift(fshift)
img_back=np.abs(np.fft.ifft2(f_ishift))
```

Display the gray scale, magnitude spectrum and the edge detection in subplot

```python
plt.subplot(131),plt.imshow(img, cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(132),plt.imshow(magnitude_spectrum, cmap = 'gray')
plt.title('Magnitude Spectrum'), plt.xticks([]), plt.yticks([])
plt.subplot(133),plt.imshow(img_back, cmap = 'gray')
plt.title('Edge Detection'), plt.xticks([]), plt.yticks([])

plt.show()
```
